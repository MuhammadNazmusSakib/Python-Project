```python
bp = sns.boxplot(da.BPXSY1)
_ = bp.set_ylabel("Blood pressure in mm/Hg")
```

```python
bp = sns.boxplot(da.loc[:, ["BPXSY1", "BPXSY2", "BPXDI1", "BPXDI2"]])
bp.set_ylabel("Blood pressure in mm/Hg")
```

```python
da["age_gap"] = pd.cut(da.RIDAGEYR, [18, 30, 40, 50, 60, 70, 80]) # Create age strata based on these cut points

plt.figure(figsize = (12, 5)) # Make boxplot of BPXSY1 stratified by age group

sns.boxplot(x = "age_gap", y = "BPXSY1", data = da)   # Make the figure wider than default (12cm wide by 5cm tall
```

```python
da["age_gap"] = pd.cut(da.RIDAGEYR, [18, 30, 40, 50, 60, 70, 80])
plt.figure(figsize = (12,5))
sns.boxplot(x = "age_gap", y = "BPXSY1", hue = "RIAGENDRx", data = da)
```
